```python
import pandas as pd
```

```python
data = pd.read_csv('data_bps2019.csv')
data.head()
```

```python
data.drop(columns='Unnamed: 0', axis=1, inplace=True)
```

```python
#add outcome column 1 for miskin and 0 tidak miskin

data['Outcome'] = data['exp_cap'].apply(lambda x: 1 if x > 327402 else 0)

data
```

```python
data['Outcome'].value_counts()
```

```python
data.to_csv('data_bps2019AddOutcome.csv', index=False)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import numpy as np
```

```python
#split data into X and Y

X = data.iloc[:, 0:8].values
Y = data.iloc[:, -1].values
```

```python
X
```

```python
Y
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3, random_state=1)
```

```python
X_train.shape
```

```python
X_test.shape
```

# Model Training

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train,Y_train)
```

# Model Evaluation

```python
X_train_pred = model.predict(X_train)
```

```python
train_data_acc = accuracy_score(X_train_pred, Y_train)

print(train_data_acc)
```

```python
X_test_pred = model.predict(X_test)
```

```python
test_data_acc = accuracy_score(X_test_pred,Y_test)

print(test_data_acc)
```

# Making a predictive system

```python
input_data = (48,1,6,1,2,1,7,1)
```

```python
# turn input_data into nummpy array

input_data_arr = np.asarray(input_data)
```
